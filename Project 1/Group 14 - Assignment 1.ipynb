## 1. Rock, Paper, Scissors 

```python
def brain_prob(count1,count2,count3):
    '''Takes the number of plays the human has made for each choice,
    and will assum the most common choice will be played by human next time.
    Computer will play the move that beats what it think the human will play.
    If the probabilty for human choice is tied between two choices, 
    comptuer will assume that the human play one play one of them and will play 
    the one that avoids a loss in either scenario (ties or wins the rouond).
    Function returns the computer choice and total count of games played.'''
    import random
    total_count = count1+count2+count3
    user = [count1,count2,count3]
    res=[0,0,0]
    for i in range(0,3):
        res[i] = (user[i]==max(user))
   
    if sum(res)==1:
        c_call = (['rock'] * res[2] + ['paper'] * res[0] + ['scissors'] * res[1])[0]
    elif sum(res)==3:
        c_call = random.choice(['rock','paper','scissors'])
    elif count1 == count2:
        c_call = 'paper'
    elif count2 == count3:
        c_call = 'scissors'  
    elif count3 == count1:
        c_call = 'rock'  

    return c_call, total_count

def result(p_call,c_call):
    '''Takes in the human choice and computer choice and will make a call on the winner.
    Prints statements telling the human if they win/lose and what the computer played.
    Also keeps track of how many wins the human has (return value).'''
    p_wins=0
    if p_call == c_call:
        print("Draw! You both chose %s!") % (p_call)
    elif p_call == 'rock':
        if c_call == "paper":
            print("You lose! %s covers %s!") % (c_call.title(), p_call)
        else:
            print("You win! %s smashes %s!") % (p_call.title(), c_call)
            p_wins += 1
    elif p_call == "paper":
        if c_call == "scissors":
            print("You lose! %s cuts %s!") % (c_call.title(), p_call)
        else:
            print("You win! %s covers %s!") % (p_call.title(), c_call)
            p_wins += 1
    elif p_call == "scissors":
        if c_call == "rock":
            print("You lose! %s smashes %s!") % (c_call.title(), p_call)
        else:
            print("You win! %s cuts %s!") % (p_call.title(), c_call)
            p_wins += 1
    return p_wins

rock_count, paper_count, scis_count = 1,1,1
p_wins = 0

'''To end the game, player can input anything that is not 'rock, paper, scissors'. 
    Computer will ask to make sure (Y/N).'''  
while True:
    c_call, total = brain_prob(rock_count, paper_count, scis_count)
    print("Rock!\nPaper!\nScissors!\n")
    p_call = raw_input("Shoot!...")
    p_call = p_call.lower()
    if p_call == 'rock':
        rock_count += 1
        p_wins += result(p_call, c_call)
    elif p_call == 'paper':
        paper_count += 1
        p_wins += result(p_call, c_call)
    elif p_call == 'scissors':
        scis_count += 1
        p_wins += result(p_call,c_call)
    else:
        ext = raw_input("What was that? Are you done playing (Y/N)?")
        ext = ext.upper()
        if ext == 'N':
            print("Okay, lets go again!")
        else:
            print("You managed to win %d time(s) out of %d game(s) played") % (p_wins, total-3)
            break
```

## 3. The Google of Quotes

```python
import re
fp = open("C:\Users\Amy Dang\OneDrive\MBSA - OneDrive\MIS S381N Data Analytics Programming\HW 1\quotes.txt", "r")
```

### A: Build a list of full quotes

```python
quo_mat = []
i = 0

'''Reach each pair of lines into one string with the quote and author and stores the strings in a list.'''
fp.seek(0)
for index, lines in enumerate(fp):
    if index%2 == 0: 
        quo_mat.append(lines.rstrip())
    if index%2 == 1:
        quo_mat[i] += (' - ' + lines.rstrip())
        i+=1
        
print((quo_mat))
```

### B: Words from full quotes

```python
quote = quo_mat[0]
res = []

def words(s):
    '''Separates each word in a quote using the regular expressions and returns a list of the words.'''
    s = s.lower()
    res = re.split('\W+',s)
    return res

print words(quote)
```

### C: Build the postings-list dictionary

```python
pos_lis_dic = {}

def freq(q, pd):
    '''Counts the frequency of each word in a quote and stores them in a dictionary.
    Each quote is used as a key in the postings-list dictionary, 
    and the corresponding value is the word and frequnecy dictionary.
    Input = quote in the positings-list dictionary.
    Returns the new postings-list dictionary with the quote added.'''
    cnt = {}
    q1 = words(q)
    for w in q1:
        if w in cnt.keys():
            cnt[w] += 1
        else:
            cnt[w] = 1
    
    if q not in pd.keys():
        pd[q] = cnt
    return pd

'''Loops through all quotes.'''
for q in quo_mat:
    freq(q,pos_lis_dic)

print pos_lis_dic
        
```

### D: Build the reverse postings-list dictionary 

```python
rev_dic = {}

'''Use the word frequencies from the posting dictionary as values and the quote as the key, and puts this in a dictionary.
Use the word as the key for the reverse dictionary.'''
for ln in quo_mat:
    quo_words = words(ln)
    for w in quo_words:
        if w in rev_dic.keys():
            rev_dic[w][ln]= pos_lis_dic[ln][w]
        else:
            rev_dic[w] = {ln: pos_lis_dic[ln][w]}

#For entire reverse postings-list dictionary
#print rev_dic

print rev_dic['entertainer']            
```

### E: Write a TF-IDF function 

```python
TF_n = 0
TF_d = 0
import math
numbers=[0]*100

def TF_IDF(word, quote):
    '''TF: Input a word and a quote and get the word frequency from the posting dictionary.
    Divide by the max word frequency for that quote. 
    IDF: numerator is total number of quotes (length of initial list in part A). 
    Denominator used the length of the quote:frequency dictionary from inside the reverse dictionary (part D).
    Take the log to get IDF.
    Returns the product of TF*IDF.'''     
    if pos_lis_dic.get(quote).get(word) is None:
        TF_n = 0
    else:
        TF_n = pos_lis_dic.get(quote).get(word)
   

   frequencies = pos_lis_dic.get(quote).items()
    for i in range(len(frequencies)):  
        numbers[i] = frequencies[i][1]
    TF_d = max(numbers)
    TF= TF_n/float(TF_d)
   
    num_quotes = len(quo_mat)
    word_quotes = len(rev_dic[word])
    IDF = math.log(num_quotes/word_quotes)
    TF_IDF = TF*IDF
   
    return TF_IDF
   
print TF_IDF('entertainer','An actor is at most a poet and at least an entertainer. - Marlon Brando')
```

### F: Quote search using a single word 

```python
def search_word(w):
    '''Input a word and find it in the reverse dictionary (word = key). 
    For each quote, corresponding to that word, use the TFIDF function from part E and returns the scores.'''    
    scores = {}
    
    for q in rev_dic.get(w).keys():
        scores[q] = TF_IDF(w,q)
    return scores

print search_word('fool')
```

### G: Quote search using multiple words 

```python
def search_multi(m):
    '''Input a list of words, find them in the reverse dictionary. 
    For each quote, calculate the store for that word and add it to the total for that quote.
    Multiple words are added together into one quote.'''    
    multi_scores = {}
    
    for w in m:
        for q in rev_dic.get(w).keys():
            if q in multi_scores.keys():
                multi_scores[q] += TF_IDF(w,q)
            else: multi_scores[q] = TF_IDF(w,q)
                          
    return multi_scores

print search_multi(['fool','nasty'])
```
